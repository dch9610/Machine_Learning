```python
# 데이터 불러오기
boston = pd.read_csv("data/Boston_house.csv")
boston.head()
```

```python
# target 제외한 데이터만 뽑기 (예측해야하는 값)
boston_data = boston.drop(['Target'], axis=1)
```

# crim/rm/lstat 세개의 변수로 각각 단순 선형 회귀 분석하기

```python
## 변수 설정 target/crim/rm/lstat
target = boston[['Target']]
crim = boston[['CRIM']]
rm = boston[['RM']]
lstat = boston[['LSTAT']]
```

## target ~ crim 선형회귀분석

```python
# crim변수에 상수항추가하기 (sm 모델 사용)

crim1 = sm.add_constant(CRIM, has_constant="add")
crim1.head()
```

```python
# sm.OLS 적합시키기 (y, 데이터)
# 최소자승법으로 회귀모델을 구함
model1 = sm.OLS(target, crim1)
fitted_model1 = model1.fit() 
```

```python
# summary함수통해 결과출력 
# y = -0.41*x + 24
# coef를 통해 범죄율이 1증가할때 y가 -0.41 증가

# pvalue를 보고 확인후 비슷하다면
# R-squared가 높은 변수가 영향이 높은 변수
fitted_model1.summary()
```

```python
## 회귀 계수 출력 (.params)
fitted_model1.params
```

### y_hat=beta0 + beta1 * X 계산해보기

```python
#회귀 계수 x 데이터(X)
pred = np.dot(crim1, fitted_model1.params)
pred
```

```python
## predict함수를 통해 yhat구하기 
pred1 = fitted_model1.predict(crim1)
pred1
```

```python
## 직접구한 yhat과 predict함수를 통해 구한 yhat차이 
# yhat은 예측값
np.dot(crim1, fitted_model1.params) - pred1
```

## 적합시킨 직선 시각화 

```python
import matplotlib.pyplot as plt

# 음의 상관관계를 가지는것을 확인
plt.yticks(fontname = "Arial") #
plt.scatter(crim,target,label="data")
plt.plot(crim,pred1,label="result")
plt.legend()
plt.show()
```

```python
plt.scatter(target,pred1)
plt.xlabel("real_value") # 실제값
plt.ylabel("pred_value") # 예측값
plt.show()
```

```python
## residual 시각화 
# 잔차가 0에서 많이 벗어난것을 확인

fitted_model1.resid.plot()
plt.xlabel("residual_number")
plt.show()
```

```python
##잔차의 합계산해보기
# 잔차(실제값-예측값)의 합은 0에 가까움
np.sum(fitted_model1.resid)
```

## 위와 동일하게 rm변수와 lstat 변수로 각각 단순선형회귀분석 적합시켜보기 

```python
# 상수항추가
rm1 = sm.add_constant(rm, has_constant="add")
lstat1 = sm.add_constant(lstat, has_constant="add")
```

```python
# 회귀모델 적합
model2 = sm.OLS(target, rm1)
fitted_model2 = model2.fit() 

model3 = sm.OLS(target, lstat1)
fitted_model3 = model3.fit() 
```

```python
# rm모델 결과 출력
# y = 9.1*x -34.7   coef(RM*x + const)
# R-squared가 가장 높은 변수가 영향이 더 높은 변수
fitted_model2.summary()
```

```python
# lstat모델 결과 출력 
# y = -1*x + 34.6
fitted_model3.summary()
```

```python
## 각각 yhat_예측하기 
pred2 = fitted_model2.predict(rm1)
pred3 = fitted_model3.predict(lstat1)
```

```python
## rm모델 시각화
import matplotlib.pyplot as plt

plt.scatter(rm,target,label="data")
plt.plot(rm,pred2,label="result")

plt.xlabel('rm')
plt.ylabel('pred2')

plt.legend()
plt.show()
```

```python
## lstat모델 직선 시각화
import matplotlib.pyplot as plt

plt.scatter(lstat,target,label="data")
plt.plot(lstat,pred3,label="result")

plt.legend()
plt.show()
```

```python
# rm모델 reisidual 시각화 
fitted_model2.resid.plot()
plt.xlabel("residual_number")
plt.show()
```

```python
##잔차의 합계산해보기
# 잔차(실제값-예측값)의 합은 0에 가까움
np.sum(fitted_model2.resid)
```

```python
# lstat모델 residual시각화 (잔차)

fitted_model3.resid.plot()
plt.xlabel("residual_number")
plt.show()
```

```python
##잔차의 합계산해보기
# 잔차(실제값-예측값)의 합은 0에 가까움
np.sum(fitted_model3.resid)
```

```python
## 새 모델의 residual비교 
# 잔차가 y=0 주변으로 몰려있는지를 판단

fitted_model1.resid.plot(label="crim")
fitted_model2.resid.plot(label="rm")
fitted_model3.resid.plot(label="lstat")
plt.legend()
```
